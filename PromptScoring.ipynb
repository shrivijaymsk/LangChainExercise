```python
llm=ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0.2)
```

```python
#("system","You are a helpful assistant, your goal is to summarize given topic"),
from langchain_core.prompts import ChatPromptTemplate

promt_template=ChatPromptTemplate.from_messages(

    [
        ("system",
         "You are a prompt quality scoring agent, your goal is to evaluate the quality of a user prompt, assign a cumulative score and provide brief feedback, improvement, suggestion. The prompt should be evaluated based on the 5 criteria mentioned. Each of the must be scored within 0-10 individually.   1. Clarity - Checks whether prompt is easy to understand and has a clear goal. 2. Specificity/Details - Evaluates whether sufficient details and improvements are provided. 3. Context - Checks if background information, audience or use case is mentioned. 4. Output format & Constraints - Checks whether expected output format, tone and length is specified. 5. Persona defined - Confirms whether a prompt assigns a specific role.   The output must contain the Final score which is the average of 5 criteria mentioned above, , feedback, improvements and concise suggestions for improvement."),
     ("user", "{InputPrompt}")
     ]
)
```

```python
from typing_extensions import Format
print ("Test 1")
prompt_value=promt_template.format_prompt(InputPrompt="You are a well known poet. Write a 3 line poem about nature")
#print (prompt_value.to_messages())
response=llm.invoke(prompt_value)
print (response.content)
```
